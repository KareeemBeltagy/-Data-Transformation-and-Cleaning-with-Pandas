```python
#import xls file and ignore unnessecary rows and columns , Renaming all the required column 
energy = pd.read_excel("Energy Indicators.xls",skiprows=16,skipfooter=38,usecols=[2,3,4,5])
energy= energy.drop(0)
energy.columns =['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy.head()
```

```python
#Replace "..."/NaN with np.Nan 
import numpy as np
energy[['Energy Supply','Energy Supply per Capita']]= energy[['Energy Supply','Energy Supply per Capita']].replace('...',np.NaN)

energy.head()
```

```python
# Convert Energy Supply to gigajoules 
energy ['Energy Supply'] = energy ['Energy Supply'].apply(lambda x : x*1000000 )  
```

```python
#creat dict. for replaced countries
import re
count_rename={"Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
}
#Remove pranthases and numbers 
energy['Country']= energy['Country'].apply(lambda x: re.sub('\([^()]*\)','',x))
energy['Country']= energy['Country'].apply(lambda x: re.sub('\d','',x))
energy['Country']= energy['Country'].str.strip()
#Rename countries names 
energy['Country']= energy['Country'].replace(count_rename)

# print(energy.iloc[197])
# print(energy.iloc[24])
```

```python
# Sciamgo Journal  Sciamgo Journal excel file.
ScimEn=pd.read_excel('scimagojr-3.xlsx')

```

```python
ScimEn.head(10)
# .info()
# energy.info()
```

```python
# joining ScimEn and energy on country column
temp_df=pd.merge(ScimEn[['Country','Rank','Citable documents','Citations','Self-citations','Citations per document','H index']], 
               energy[['Country','Energy Supply','Energy Supply per Capita','% Renewable']],
                how='inner',on ='Country')
```

```python
temp_df.head(15)
```

```python
all_df= pd.merge(temp_df,GPD[['Country Name','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']],
                              how ='inner',left_on='Country',right_on='Country Name')
all_df=all_df.drop(['Country Name'],axis=1)
all_df=all_df.sort_values('Rank')
```

```python
#keeping the hights 15 countries in Rank 
all_df=all_df[all_df.Rank <=15]
all_df.set_index('Country')
```

```python
# number of rows lost after reduction 
#number of rows in 'ScimEn'(left table) minus number of rows after reduction 
red_rows=len(ScimEn)-len(all_df)
red_rows
```

```python
#average GPD per country for last 10 years
avg_years=['2006','2007','2008','2008','2009','2010','2011','2012','2013','2014','2015']
avgGPD=list(all_df[avg_years].mean(axis=1,skipna=True,numeric_only=None))
country=list(all_df.loc[:,'Country'])
avgGPD= list(zip(country,avgGPD))
avgGPD = sorted(avgGPD,key=lambda x: x[1], reverse=True)
display(avgGPD)
```

```python
avg_energySupplyPerCaptia=all_df['Energy Supply per Capita'].mean()
print('Average energy supply per captia = ',avg_energySupplyPerCaptia)
```

```python
# max renewable country
all_df['% Renewable']=all_df['% Renewable'].astype(float) #change type of %renwable to use .idxmax function 
max_indx=all_df['% Renewable'].idxmax()                   #get index of max country
#retrive name of country and its vlaue 
max_ren=(all_df.loc[max_indx,'Country'],all_df.loc[max_indx,'% Renewable'])
max_ren
```

```python
#calculate ratio of self citation to total citaion
self_to_total_citaion=list(all_df['Self-citations']/all_df['Citations'])
#insert the list as new column to dataframe
all_df.insert(7,'Self to total citation',self_to_total_citaion)
```

```python
# get max value in 'Self to total citation' column.
max_indx=all_df['Self to total citation'].idxmax()
max_cit=(all_df.loc[max_indx,'Country'],all_df.loc[max_indx,'Self to total citation'])
print('max country of Self to total citation is ',max_cit)
```

```python
#calculate meadian of Renewable 
median_renewable = all_df['% Renewable'].median()
#create a list of 1 for country higher than median else 0
highrenew = list((all_df['% Renewable'] >= median_renewable).astype(int))
#insert list as new column 
all_df.insert(11,'HighRenew',highrenew)
#sort dataframe by rank
all_df = all_df.sort_values('Rank')
all_df.style
# highrenew
```

```python
# create HighRenew Series
HighRenew=pd.Series(all_df['HighRenew'].values, index=all_df['Country'])
HighRenew
```

```python
# create list of population
popuList=list(all_df['Energy Supply']/all_df['Energy Supply per Capita'])
popuList
# continent country dict.
ContinentDict  = {'China':'Asia',  'United States':'North America', 
                  'Japan':'Asia', 'United Kingdom':'Europe', 
                  'Russian Federation':'Europe',  'Canada':'North America',  'Germany':'Europe', 
                  'India':'Asia','France':'Europe', 
                  'South Korea':'Asia',  'Italy':'Europe', 
                  'Spain':'Europe', 'Iran':'Asia',
                  'Australia':'Australia',  'Brazil':'South America'}
ContinentLSeries=pd.Series(ContinentDict)
ContinentLSeries
Continent_df=pd.DataFrame({'Continent': ContinentLSeries.values , 'Country': ContinentLSeries.index}).set_index('Continent')
# .set_index('Continent')

#insert population column
Continent_df.insert(1,'Population',popuList)

Continent_df.info()
```

```python
import numpy as np
AggContinent_df=Continent_df.groupby('Continent')['Population'].agg([np.size,np.sum,np.mean,np.std])

AggContinent_df
```
